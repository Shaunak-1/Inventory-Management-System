```python
import json
import time

fd=open("records.json",'r')
js=fd.read()
fd.close()

sale=''
record= json.loads(js)
print("-----------------MENU-------------------")
for key in record.keys():
    print(key,":",record[key]["Name"],"\t|",record[key]["Price"],"\t|",record[key]["Qn"])
    
print("----------------------------------------\n")

ui_pr=str(input("Enter product id: "))
ui_qn=int(input("Enter quantity  : "))
ui_name=str(input("Enter your name :"))
ui_id=str(input("Enter your email Id :"))

if(record[ui_pr]["Qn"]>=ui_qn):
    record[ui_pr]["Qn"]=record[ui_pr]["Qn"]-ui_qn
    print("-----------------------------------------")
    print("                  Bill                   \n")
    print("Name           : ",record[ui_pr]["Name"])
    print("Price          : ",record[ui_pr]["Price"])
    print("Quantity       : ",ui_qn)
    print("-----------------------------------------")

    print("Billing amount : ",record[ui_pr]["Price"]*ui_qn)

    print("-----------------------------------------")
    sale=ui_name+","+ui_id+','+record[ui_pr]['Name']+','+str(record[ui_pr]['Price'])+','+str(ui_qn)+','+str(record[ui_pr]['Price']*ui_qn)+','+time.ctime()+'\n'
else:
    print("We dont have enough quantity.")
    print("We have only",record[ui_pr]["Qn"],"quantity")
    ch=input("Press Y/y to purchase :")
    if(ch=='Y' or ch=='y'):
        print("-----------------------------------------")
        print("                  Bill                   \n")
        print("Name           : ",record[ui_pr]["Name"])
        print("Price          : ",record[ui_pr]["Price"])
        print("Quantity       : ",record[ui_pr]["Qn"])
        print("-----------------------------------------")

        print("Billing amount : ",record[ui_pr]["Price"]*record[ui_pr]["Qn"])

        print("-----------------------------------------")
        sale=ui_name+","+ui_id+','+record[ui_pr]['Name']+','+str(record[ui_pr]['Price'])+','+str(record[ui_pr]["Qn"])+','+str(record[ui_pr]['Price']*record[ui_pr]["Qn"])+','+time.ctime()+'\n'
        record[ui_pr]["Qn"]=0
        
    else:
        print("Thanks!")

js=json.dumps(record)
fd=open("records.json",'w')
fd.write(js)
fd.close()

fd=open('sales.txt','a')
fd.write(sale)
fd.close()
```
